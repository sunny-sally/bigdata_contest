## 분석 위해 필요한 열만 생각해보기
- 상호명, 지점명, 상권업종대분류명, 상권업종중분류명, 상권업종소분류명, 도로명주소, 층정보, 법정동명, 경도, 위도

## 열 사용 방안
- 중분류로 업종 나누기
- 주소 간단히 합치기
- 가게 대표 서비스는 소분류명으로 생각
- 상호명, 지점명 합치기

```python
df_gwangjin  = df_g[['상호명', '지점명', '상권업종대분류명', '상권업종중분류명', '상권업종소분류명', '도로명주소','법정동명', '층정보', '경도', '위도']]
df_gwangjin
```

```python
print(sum(df_gwangjin['층정보'].isnull()))
```

```python
# 주소 간단히 합치기

df_gwangjin = pd.DataFrame(df_gwangjin)

def address(x):
    if pd.notnull(x['층정보']):  # '층정보' 열이 NaN이 아닐 경우
        return x['도로명주소'] + " (" + x['법정동명'] + ") " + str(x['층정보']) + "층"
    else:  # '층정보' 열이 NaN일 경우
        return x['도로명주소'] + " (" + x['법정동명'] + ")"


# '주소' 열 생성
df_gwangjin['주소'] = df_gwangjin.apply(address, axis=1)

# 결과 확인
print(df_gwangjin[['주소']])
```

```python
# 상호, 지점명 합치기
def name(x):
    if pd.notnull(x['지점명']):
        return x['상호명'] + " " + x['지점명']
    else:  # '층정보' 열이 NaN일 경우
        return x['상호명']

df_gwangjin['상호'] = df_gwangjin.apply(name, axis=1)
```

```python
df_gwangjin
```

```python
# '상호명'에 '점'이 포함된 행만 필터링하여 보기
df_gwangjin_with_점 = df_gwangjin[df_gwangjin['상호명'].str.contains('점', na=False)]

# 결과 확인
print(df_gwangjin_with_점[['상호명']])

```

- 이미 상호명에 지점명이 들어간 경우도 있음 <- 어쩔수 없을듯

```python
# 다 쓴 도로명주소, 법정동명, 층정보 삭제
df_gwangjin = df_gwangjin.drop(['법정동명', '층정보', '상호명', '지점명'], axis=1)
df_gwangjin
```

```python
# 열 순서 변경
new_order = ['상호', '주소', '상권업종대분류명', '상권업종중분류명', '상권업종소분류명', '경도', '위도', '도로명주소']

df_gwangjin = df_gwangjin[new_order]
df_gwangjin
```
