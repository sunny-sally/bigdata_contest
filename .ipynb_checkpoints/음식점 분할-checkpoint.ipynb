```python
import pandas as pd
```

```python
df_food = pd.read_csv('data/gwangjin_food.csv')
df_food
```

```python
hygiene_df = pd.read_csv('data/μ„μƒλ“±κΈ‰μ—…μ†_μ „μ²λ¦¬.csv')
hygiene_df
```

```python
import re
# π§Ή μ£Όμ† μ •λ¦¬ ν•¨μ
def clean_address(address):
    if pd.isna(address):
        return address
    # κ΄„νΈ λ° κ΄„νΈ μ• λ‚΄μ© μ κ±°
    cleaned_address = re.sub(r'\s*\(.*?\)', '', address)  # κ΄„νΈ μ κ±°
    # μΈµ μ •λ³΄ μ κ±° (μ«μ+μΈµ ν¨ν„΄)
    cleaned_address = re.sub(r'\s*\d+μΈµ', '', cleaned_address)  # μΈµ μ •λ³΄ μ κ±°
    # λ¶ν•„μ”ν• νΉμ λ¬Έμλ‚ λΉ„μ •μƒμ μΈ λ¬Έμ μ κ±°
    cleaned_address = re.sub(r'[^a-zA-Z0-9κ°€-ν£\s\-.,]', '', cleaned_address)  # μ •μƒμ μΈ λ¬Έμλ§ λ‚¨κΈ°κΈ°
    return cleaned_address.strip()  # μ–‘ λ κ³µλ°± μ κ±°

# π—οΈ μ£Όμ† μ •λ¦¬ μ μ©
hygiene_df['μ£Όμ†μ •λ¦¬'] = hygiene_df['μ†μ¬μ§€(λ„λ΅λ…)'].apply(clean_address)

# κ²°κ³Ό ν™•μΈ
print(hygiene_df[['μ†μ¬μ§€(λ„λ΅λ…)', 'μ£Όμ†μ •λ¦¬']])
```

```python
hygiene_df
```

```python
# μ£Όμ† + μ—…μ†λ…
filtered_df = pd.merge(hygiene_df, df_food, 
                       left_on=['μ£Όμ†μ •λ¦¬', 'μ—…μ†λ…'], 
                       right_on=['λ„λ΅λ…μ£Όμ†', 'μƒνΈ'], 
                       how='inner')

# μΈλ±μ¤ λ¦¬μ…‹
filtered_df = filtered_df.reset_index(drop=True)

# κ²°κ³Ό μ¶λ ¥
print(filtered_df)
```

```python
import pandas as pd
import re
from fuzzywuzzy import fuzz

# π“ μ—…μΆ… ν‚¤μ›λ“ μ‚¬μ „ (λΉ„κµμ—μ„ μ μ™Έν•  λ‹¨μ–΄λ“¤)
ignore_words = ['μ¤νƒ€μ‹ν‹°μ ', 'λ΅―λ°', 'λ΅―λ°κ±΄λ€μ ', 'λ΅―λ°μ¤νƒ€μ‹ν‹°μ ', 'λ΅―λ°κ±΄λ€μ¤νƒ€μ‹ν‹°','μ–΄λ¦°μ΄λ€κ³µμ›μ ','λ΅―λ°κ±΄λ€','κ±΄λ€μ¤νƒ€μ‹ν‹°','κ±΄λ€μ¤νƒ€']

# π“ μƒνΈλ… μ •μ  ν•¨μ: νΉμ • λ‹¨μ–΄ μ κ±°
def clean_name(name):
    pattern = '|'.join(ignore_words)  # μ •κ·μ‹ ν¨ν„΄ μƒμ„±
    return re.sub(pattern, '', name).strip()  # νΉμ • λ‹¨μ–΄ μ κ±° ν›„ κ³µλ°± μ •λ¦¬

# π“ λ§¤μΉ­ κ²°κ³Ό μ €μ¥ λ¦¬μ¤νΈ
matched_rows = []

# π“ μ μ‚¬λ„ κΈ°μ¤€ μ„¤μ • (μ μ‚¬λ„ 60 μ΄μƒμ΄λ©΄ κ°™μ€ μ—…μ†λ΅ κ°„μ£Ό)
similarity_threshold = 60

# π“ μ£Όμ†μ™€ μƒνΈ λΉ„κµ
for i, row in df_food.iterrows():
    # λ„λ΅λ…μ£Όμ†κ°€ κ°™μ€ ν–‰ ν•„ν„°λ§
    matched_df2 = hygiene_df[hygiene_df['μ£Όμ†μ •λ¦¬'] == row['λ„λ΅λ…μ£Όμ†']]

    # μƒνΈλ… μ „μ²λ¦¬ (λΉ„κµν•  μƒνΈλ… ν΄λ¦°)
    cleaned_name_df = clean_name(row['μƒνΈ'])

    for _, row2 in matched_df2.iterrows():
        cleaned_name_df2 = clean_name(row2['μ—…μ†λ…'])

        # 1. N-gram κΈ°λ° μ μ‚¬λ„ (λ¶€λ¶„ λ¬Έμμ—΄ λΉ„κµ)
        bigram_similarity = fuzz.partial_ratio(cleaned_name_df, cleaned_name_df2)

        # 2. μ „μ²΄ λ¬Έμμ—΄ μ μ‚¬λ„
        full_similarity = fuzz.ratio(cleaned_name_df, cleaned_name_df2)

        # 3. μµμΆ… μ μ‚¬λ„ (λ‘ κ° ν‰κ· κ°’ μ‚¬μ©)
        final_similarity = (bigram_similarity + full_similarity) / 2

        # π“ μ μ‚¬λ„κ°€ κΈ°μ¤€ μ΄μƒμ΄λ©΄ λ§¤μΉ­ λ¦¬μ¤νΈμ— μ¶”κ°€
        if final_similarity >= similarity_threshold:
            matched_rows.append({
                'μƒνΈ': row['μƒνΈ'],
                'λ„λ΅λ…μ£Όμ†': row['λ„λ΅λ…μ£Όμ†'],
                'μ—…μ†λ…': row2['μ—…μ†λ…'],
                'μ£Όμ†μ •λ¦¬': row2['μ£Όμ†μ •λ¦¬'],
                'μ μ‚¬λ„': final_similarity  # μ μ‚¬λ„ ν™•μΈμ©
            })

# π“ κ²°κ³Ό DataFrame μƒμ„± λ° μ •λ ¬
matched_df = pd.DataFrame(matched_rows)

if not matched_df.empty:
    matched_df['μ—…μ†λ…'] = pd.Categorical(matched_df['μ—…μ†λ…'], categories=hygiene_df['μ—…μ†λ…'], ordered=True)
    matched_df = matched_df.sort_values('μ—…μ†λ…').reset_index(drop=True)

# π“ κ²°κ³Ό ν™•μΈ
print(matched_df)
```
