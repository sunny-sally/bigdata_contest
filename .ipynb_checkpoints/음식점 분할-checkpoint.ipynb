```python
import pandas as pd
```

```python
df_food = pd.read_csv('data/gwangjin_food.csv')
df_food
```

```python
hygiene_df = pd.read_csv('data/위생등급업소_전처리.csv')
hygiene_df
```

```python
import re
# 🧹 주소 정리 함수
def clean_address(address):
    if pd.isna(address):
        return address
    # 괄호 및 괄호 안 내용 제거
    cleaned_address = re.sub(r'\s*\(.*?\)', '', address)  # 괄호 제거
    # 층 정보 제거 (숫자+층 패턴)
    cleaned_address = re.sub(r'\s*\d+층', '', cleaned_address)  # 층 정보 제거
    # 불필요한 특수 문자나 비정상적인 문자 제거
    cleaned_address = re.sub(r'[^a-zA-Z0-9가-힣\s\-.,]', '', cleaned_address)  # 정상적인 문자만 남기기
    return cleaned_address.strip()  # 양 끝 공백 제거

# 🏗️ 주소 정리 적용
hygiene_df['주소정리'] = hygiene_df['소재지(도로명)'].apply(clean_address)

# 결과 확인
print(hygiene_df[['소재지(도로명)', '주소정리']])
```

```python
hygiene_df
```

```python
# 주소 + 업소명
filtered_df = pd.merge(hygiene_df, df_food, 
                       left_on=['주소정리', '업소명'], 
                       right_on=['도로명주소', '상호'], 
                       how='inner')

# 인덱스 리셋
filtered_df = filtered_df.reset_index(drop=True)

# 결과 출력
print(filtered_df)
```

```python
import pandas as pd
import re
from fuzzywuzzy import fuzz

# 📌 업종 키워드 사전 (비교에서 제외할 단어들)
ignore_words = ['스타시티점', '롯데', '롯데건대점', '롯데스타시티점', '롯데건대스타시티','어린이대공원점','롯데건대','건대스타시티','건대스타']

# 📌 상호명 정제 함수: 특정 단어 제거
def clean_name(name):
    pattern = '|'.join(ignore_words)  # 정규식 패턴 생성
    return re.sub(pattern, '', name).strip()  # 특정 단어 제거 후 공백 정리

# 📌 매칭 결과 저장 리스트
matched_rows = []

# 📌 유사도 기준 설정 (유사도 60 이상이면 같은 업소로 간주)
similarity_threshold = 60

# 📊 주소와 상호 비교
for i, row in df_food.iterrows():
    # 도로명주소가 같은 행 필터링
    matched_df2 = hygiene_df[hygiene_df['주소정리'] == row['도로명주소']]

    # 상호명 전처리 (비교할 상호명 클린)
    cleaned_name_df = clean_name(row['상호'])

    for _, row2 in matched_df2.iterrows():
        cleaned_name_df2 = clean_name(row2['업소명'])

        # 1. N-gram 기반 유사도 (부분 문자열 비교)
        bigram_similarity = fuzz.partial_ratio(cleaned_name_df, cleaned_name_df2)

        # 2. 전체 문자열 유사도
        full_similarity = fuzz.ratio(cleaned_name_df, cleaned_name_df2)

        # 3. 최종 유사도 (두 개 평균값 사용)
        final_similarity = (bigram_similarity + full_similarity) / 2

        # 📌 유사도가 기준 이상이면 매칭 리스트에 추가
        if final_similarity >= similarity_threshold:
            matched_rows.append({
                '상호': row['상호'],
                '도로명주소': row['도로명주소'],
                '업소명': row2['업소명'],
                '주소정리': row2['주소정리'],
                '유사도': final_similarity  # 유사도 확인용
            })

# 📌 결과 DataFrame 생성 및 정렬
matched_df = pd.DataFrame(matched_rows)

if not matched_df.empty:
    matched_df['업소명'] = pd.Categorical(matched_df['업소명'], categories=hygiene_df['업소명'], ordered=True)
    matched_df = matched_df.sort_values('업소명').reset_index(drop=True)

# 📊 결과 확인
print(matched_df)
```

```python
# 출력할 최대 행 수 설정
pd.set_option('display.max_rows', None)  # 모든 행을 출력

# 출력할 최대 열 수 설정
pd.set_option('display.max_columns', None)  # 모든 열을 출력

# 이제 matched_df의 모든 데이터를 출력
print(matched_df)
```
