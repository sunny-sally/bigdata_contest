```python
file_path = r"C:\Users\sally\OneDrive\바탕 화면\25_1\guangginGu_project\착한가격업소_주소전처리ver.csv"

df_good = pd.read_csv(file_path, encoding="utf-8")  # 'utf-8' 인코딩 사용

print(df_good.head())  # 첫 5행 출력
```

```python
## sql 사용해보기
import sqlite3

conn = sqlite3.connect("guangginGu_data.db")

# 커서 생성
cursor = conn.cursor()

# 테이블 삭제
cursor.execute("DROP TABLE IF EXISTS stores")

# 변경사항 저장
conn.commit()
```

```python
df_gwangjin.to_sql("stores",conn,if_exists="replace",index = False)
```

```python
result = pd.read_sql("SELECT * FROM stores Where 상호명='자이헤어'",conn)
print(result)
```

```python
conn.close()
```

```python
#다시 추출 시도
from tabulate import tabulate

df_merged = df_gwangjin.merge(df_good, left_on='도로명주소', right_on='주소_추출', how='inner')

df_result = df_merged[['도로명주소', '상호명']].rename(columns={'도로명주소': '도로명 주소', '상호명': '상호명'}).sort_values(by='도로명 주소').reset_index(drop=True)

print(tabulate(df_result, headers='keys', tablefmt='pretty'))
```

```python
df_matched = df_gwangjin.merge(df_good, left_on=['도로명주소', '상호명'], right_on=['주소_추출','업소명'],how='inner')

# 결과 출력
print(df_matched[['도로명주소', '상호명']])
```
